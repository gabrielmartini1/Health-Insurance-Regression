## **4. Divisão dos dados (data splitting)**

<p align="justify">Data splitting, também conhecido como divisão de dados, é uma técnica usada em machine learning para dividir o conjunto de dados disponível em conjuntos de treinamento e teste. O conjunto de treinamento é usado para treinar o modelo de machine learning, enquanto o conjunto de teste é usado para avaliar a capacidade de generalização do modelo para novos dados não vistos durante o treinamento.</p>

<p align="justify">A técnica de divisão de dados é fundamental em machine learning para evitar o overfitting, que ocorre quando o modelo é treinado em excesso com o conjunto de treinamento e perde a capacidade de generalização para novos dados. A divisão dos dados permite que o modelo seja treinado em um conjunto de dados independente do conjunto de teste, o que ajuda a avaliar a capacidade de generalização do modelo.</p>

<p align="justify">Os conjuntos de treinamento e teste são geralmente divididos em proporções comuns, como 70-30, 80-20 ou 90-10, em que a maioria dos dados é usada para treinar o modelo (conjunto de treinamento) e uma pequena parcela é usada para testar a capacidade de generalização do modelo (conjunto de teste). É importante garantir que a divisão dos dados seja feita de forma aleatória e estratificada, para garantir que as classes ou distribuições de características do conjunto de dados sejam mantidas em ambas as partições.</p>

```python
# Separando os dados em teste e treino
test = df.sample(frac=0.10)
train = df.drop(test.index)

test.reset_index(inplace=True, drop=True)
train.reset_index(inplace=True, drop=True)
```

<p align="justify">Conforme o código apresentado acima, já efetuamos a separação dos dados em teste e treino, aplicando uma amostra aleatória fracionada em 10% para o conjunto de teste.</p>

```python
# Verificando o formato dos conjuntos
print(train.shape)
print(test.shape)
```

<p align="justify">Através do comando "shape( )", podemos  verificar a integridade e tamanho dos conjuntos de treino e teste. Observa-se, que o conjunto de treino do DataFrame apresenta 1.203 linhas e 7 colunas, enquanto que o conjunto de teste apresenta 134 linhas e 7 colunas.</p>


## **5. Criando o ambiente de modelagem no PyCaret**

<p align="justify">A criação de um ambiente de modelagem no PyCaret é importante por várias razões. Em primeiro lugar, o PyCaret é uma biblioteca de machine learning de alto nível que permite o desenvolvimento e a avalição de modelos de machine learning rapidamente, sem ter que escrever muito código. No entanto, para aproveitar ao máximo o PyCaret, é necessário criarmos um ambiente de modelagem adequado usando a função setup( ).</p>

<p align="justify">Ao criarmos um ambiente de modelagem no PyCaret, garantimos que os dados sejam pré-processados adequadamente para modelagem, que as variáveis independentes e dependentes estejam corretamente definidas, que a divisão de dados seja feita de forma aleatória e estratificada e que o ambiente de modelagem seja configurado corretamente com as bibliotecas e dependências necessárias.</p>

<p align="justify">Além disso, ao criarmos um ambiente de modelagem no PyCaret, podemos simplificar muitas das longas tarefas associadas ao pré-processamento de dados, como o tratamento de valores ausentes, a codificação de variáveis categóricas e normalização de dados. ### **5.1 Criando múltiplas tarefas com a Função Setup**

<p align="justify">Como já sabemos, a criação do ambiente de modelagem no PyCaret é realizada através da função "setup( )", a qual realiza várias tarefas importantes para preparar os dados para a modelagem, incluindo:</p>

* <p align="justify">Carregar os dados: a função "setup( )" carrega o conjunto de dados no ambiente PyCaret para que possa ser usado em análises e modelagem de machine learning.</p>

* <p align="justify">Definir a variável target: precisamos especificar qual coluna do conjunto de dados é a variável target para o modelo de Machine Learning.</p>

* <p align="justify">Selecionar as variáveis independentes: precisamos especificar quais colunas do conjunto de dados são as variáveis independentes. Essas são as variáveis que o modelo usará para prever a variável target.</p>

* <p align="justify">Pré-processar os dados: a função setup( ) aplica automaticamente várias transformações de pré-processamento de dados, incluindo tratamento de valores ausentes, codificação de variáveis categóricas, normalização de dados e outras técnicas de pré-processamento. Essas transformações ajudam a garantir que os dados estejam em um formato adequado para a modelagem.</p>

* <p align="justify">Configurar o ambiente: a função setup( ) configura automaticamente o ambiente PyCaret com as bibliotecas e dependências necessárias para a modelagem de machine learning.</p>

* <p align="justify">O resultado da função setup( ) é um objeto que contém as informações sobre as transformações de pré-processamento aplicadas aos dados, bem como os conjuntos de treinamento e teste, que são usados na etapa de treinamento do modelo. Com isso, podemos passar esse objeto para as funções de modelagem do PyCaret para avaliarmos diferentes modelos de machine learning.</p>

```python
# Configurando o ambiente de modelagem
reg = setup(data = train, target = 'charges', session_id=1) 
```

<p align="justify">Quando o código da função "setup( )" é executado corretamente, o ambiente do PyCaret imprime um grid com várias informações importantes sobre o pipeline de pré-processamento.</p> 

<p align="justify">Ao defirnirmos o parâmetro "session_id=1" na função "setup( )", estamos garantindo que o processamento dos dados e a construção do modelo serão executados de forma determinística, ou seja, com a mesma aleatoriedade a cada execução do código.</p>

<p align="justify">O valor que é definido para o parâmetro "session_id=1", estabelece uma semente ("seeding") para a aleatoriedade, o que garante que o processamento das linhas de código seja sempre realizado da mesma forma. Com isso, a reprodutibilidade do processo de modelagem é garantida, o que é importante para avaliarmos a performance do modelo de forma consistente e confiável.</p>

Isso pode acontecer, porque a busca em grade não encontrou nenhuma combinação de hiperparâmetros que melhorasse significativamente o desempenho do modelo em relação ao original.</p>

<p align="justify">É sempre muito importante avaliarmos cuidadosamente o desempenho do modelo ajustado em comparação com o modelo original, bem como interpretar os resultados da busca em grade com cuidado para entendermos como os hiperparâmetros afetam o desempenho do modelo.</p>

* <p align="justify">Desenvolvimento de produtos: O modelo GBR pode ajudar as seguradoras a identificar os recursos e benefícios mais valorizados pelos clientes, permitindo que desenvolvam produtos que atendam melhor às necessidades dos clientes.</p>
